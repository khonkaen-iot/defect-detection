```python
import cv2
```

```python
image = results[0].orig_img.copy
```

```python
cv2.putText(image, "This is Screw", (30,100),cv2.FONT_HERSHEY_SIMPLEX,2, (0,255,0),2)
plt.imshow(image)
```

```python
bbox = np.array(results[0].boxes.xyxy.cpu(), dtype='int')[0]
bbox
```

```python
preds_name = classnames[cls[0]]
x1,x2,y1,y2 = bbox
cv2.rectangle(image, (x1,y1),(x2,y2),(255,0,0),3)
cv2.putText(image, preds_name, (x1,y1-10),cv2.FONT_HERSHEY_SIMPLEX,2, (255,0,0),2)
plt.imshow(image)

```
