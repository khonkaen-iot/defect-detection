```python
import cv2
import matplotlib.pyplot as plt


def plot_predict(image_path):
  '''
  image_path : str (path of image file that need
  to be predicted)
  '''

  image = cv2.imread(image_path)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

  results = model(image_path)
  result = results[0]
  bboxes = np.array(result.boxes.xyxy.cpu(), dtype='int')
  classes = np.array(result.boxes.cls.cpu(),dtype='int')
  confidences = np.array(result.boxes.conf.cpu())
  class_names = result.names
  class_name_list = [class_names[i] for i in classes]

  conf_threshold = 0

  for cls, bbox , conf in zip(classes,bboxes, confidences):
    if conf > conf_threshold:
      x1,y1,x2,y2 = bbox
      label = class_names[cls]
      cv2.rectangle(image, (x1,y1),(x2,y2),(255,0,0),3)
      cv2.putText(image, label, (x1,y1-10),cv2.FONT_HERSHEY_SIMPLEX,2, (255,0,0),2)


  plt.imshow(image)
  plt.pause(0.1)
```

```python
plot_predict('/content/screw/test/scratch_head/003.png')
```

```python
import os
folder_path = '/content/screw/test/manipulated_front'

for filename in os.listdir(folder_path):
    if filename.endswith(('.jpg', '.jpeg', '.png')):
        image_path = os.path.join(folder_path, filename)
        plot_predict(image_path)
```
